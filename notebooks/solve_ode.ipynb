To solve a simple ODE numerically in Python, you can use the solve_ivp function from the scipy.integrate module. Here's an example:
This code solves the ODE dy/dt = -2y with the initial condition y(0) = 1 and plots the solution over the interval.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE
def dydt(t, y):
    return -2 * y

# Initial condition
y0 = [1]

# Time points
t_span = (0, 5)

# Solve the ODE
sol = solve_ivp(dydt, t_span, y0)

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('y')
plt.title('Numerical Solution of dy/dt = -2y')
plt.show()
```

To solve a simple ODE numerically in Python, you can use the solve_ivp function from the scipy.integrate module. Here's an example:
This code solves the ODE dy/dt = 5y with the initial condition y(0) = 2 and plots the solution over the interval.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE
def dydt(t, y):
    return 5 * y

# Initial condition
y0 = [0.25]

# Time points
t_span = (0, 3)

# Solve the ODE
sol = solve_ivp(dydt, t_span, y0)

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('y')
plt.title('Numerical Solution of dy/dt = 5y')
plt.show()
```
