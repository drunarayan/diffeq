## Bunnies differential Equation Solver

Explanation:
1. Import Libraries:
numpy is used for numerical operations.
matplotlib.pyplot is used for plotting.
2. euler_method Function:
Takes the function f defining the ODE, initial values x0 and y0, step size h, and final x-value xf as input.
Initializes lists for x and y values.
Iteratively calculates the next x and y values using Euler's method until xf is reached.
Returns the x and y values as lists.
3. Example Usage:
Defines the ODE y' = y using the function f.
Sets initial values, step size, and final x-value.
Calls the euler_method function to solve the ODE.
Plots the solution using matplotlib.pyplot.

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, lam, x0, y0, h, xf):
    """Solves the ODE y' = f(x, y) using the Euler method."""
    x = [x0]
    y = [y0]

    while x[-1] < xf:
        x_next = x[-1] + h
        y_next = y[-1] + h * f(x[-1], y[-1],lam)
        x.append(x_next)
        y.append(y_next)

    return x, y

# Differential Equation
def f(x, y, lam):
    return lam * y

x0 = 0
y0 = 10
h = .01
xf = 1
lam = 5

x, y = euler_method(f, lam, x0, y0, h, xf)

plt.plot(x, y)
# plot exact solution
x = np.linspace(0,1)
plt.plot(x, np.exp(lam*x)*y0, 'k--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution to dy/dt = lam * y using Euler method')
plt.show()
```
